```python
#creating a function 
from datetime import datetime
def convert_datetime(dt):
    in_time = datetime.strptime(dt, "%Y-%m-%d %H:%M:%S")
    return datetime.strftime(in_time, "%m/%d/%Y %H:%M:%S")
```

```python
Green_trip['pickup_time'] = pd.to_datetime(Green_trip['pickup_time'].map(convert_datetime))
Green_trip['dropoff_time'] = pd.to_datetime(Green_trip['dropoff_time'].map(convert_datetime))
```

```python
#removing the dropoff time greater than pickup time values from dataset.
```

```python
Green_trip = Green_trip[Green_trip["dropoff_time"] > Green_trip['pickup_time']]
```

```python
Green_trip.index = pd.DatetimeIndex(Green_trip['pickup_time'])
```

```python
# trip distance > 0
data = Green_trip[Green_trip['trip_distance'] > 0]
trip_dist = Green_trip['trip_distance']
Q1 = trip_dist.quantile(0.25)
Q3 = trip_dist.quantile(0.75)
IQR = Q3 - Q1
Green_trip = data[-((data['trip_distance'] < (Q1 - 1.5 * IQR)) | (data['trip_distance'] > (Q3 + 1.5 * IQR)))]
```

```python
# tip amount >= 0
data = Green_trip[Green_trip['tip_amount'] >= 0]
tip_amt = Green_trip['tip_amount']
Q1 = tip_amt.quantile(0.25)
Q3 = tip_amt.quantile(0.75)
IQR = Q3 - Q1
Green_trip = data[-((data['tip_amount'] < (Q1 - 1.5 * IQR)) | (data['tip_amount'] > (Q3 + 1.5 * IQR)))]
```

```python
# total amount >= 2.5
data = Green_trip[Green_trip['total_amount'] >= 2.5]
total_amt = Green_trip['total_amount']
Q1 = total_amt.quantile(0.25)
Q3 = total_amt.quantile(0.75)
IQR = Q3 - Q1
Green_trip = data[-((data['total_amount'] < (Q1 - 1.5 * IQR)) | (data['total_amount'] > (Q3 + 1.5 * IQR)))]
```

```python
# fare amount >= 2.5
data = Green_trip[Green_trip['fare_amount'] >= 2.5]
fare_amt = Green_trip['fare_amount']
Q1 = fare_amt.quantile(0.25)
Q3 = fare_amt.quantile(0.75)
IQR = Q3 - Q1
Green_trip = data[-((data['fare_amount'] < (Q1 - 1.5 * IQR)) | (data['fare_amount'] > (Q3 + 1.5 * IQR)))]
```

```python
# Calc. Trip Duration - in Minutes
Green_trip['trip_duration'] = (Green_trip['dropoff_time'] - Green_trip['pickup_time']).dt.total_seconds()/60
duration = Green_trip['trip_duration']
Q1 = duration.quantile(0.25)
Q3 = duration.quantile(0.75)
IQR = Q3 - Q1
Green_trip = Green_trip[-((Green_trip['trip_duration'] < (Q1 - 1.5 * IQR)) | (Green_trip['trip_duration'] > (Q3 + 1.5 * IQR)))]
```

```python
Green_trip['trip_avg_speed'] = Green_trip['trip_distance'] / (Green_trip['trip_duration']/60) #miles/hour data['trip_avg_speed']
```

```python
np.random.seed(42)
samples = data['trip_distance']
shape, loc, scale = stats.lognorm.fit(samples, floc=0)
x_fit = np.linspace(samples.min(), samples.max(), 100)
samples_fit = stats.lognorm.pdf(x_fit, shape, loc=loc, scale=scale)


sns.histplot(data=data, x='trip_distance', bins=20, kde=False, label='histogram')

area_hist = len(samples) * (samples.max() - samples.min()) / 20 # Calculate the area of the histogram bars

plt.plot(x_fit, samples_fit * area_hist, linewidth=2)

plt.xlabel('Trip Distance (Miles)')
plt.ylabel('Number of Rides')
plt.show()
```
